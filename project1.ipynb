<h1 > Week 2 Tasks- Sql and attribution using sqlite(maintaining sql like file with extension .sql</h1>

<h1>Create Database Connection</h1>

```python
import sqlite3

conn = sqlite3.connect("marketing.db")

print("Database Created")
```

<h1>Check Existing Tables</h1>

```python
import pandas as pd
cursor = conn.cursor()
query = """
SELECT name
FROM sqlite_master
WHERE type='table'
"""

pd.read_sql(query, conn)
```

<h1>Create Ad Spend Table</h1>

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS ad_spend (
    spend_id INTEGER,
    date TEXT,
    channel TEXT,
    campaign TEXT,
    utm_source TEXT,
    utm_medium TEXT,
    utm_campaign TEXT,
    impressions INTEGER,
    clicks INTEGER,
    ad_spend REAL,
    country TEXT
)
""")

conn.commit()

print("Ad Spend Table Created")
```

<h1>Create Web Analytics Table</h1>

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS web_logs (
    session_id TEXT,
    user_id TEXT,
    session_date TEXT,
    traffic_source TEXT,
    utm_source TEXT,
    utm_medium TEXT,
    utm_campaign TEXT,
    device TEXT,
    city TEXT,
    country TEXT,
    pages_viewed INTEGER,
    session_duration INTEGER,
    bounce_flag INTEGER
)
""")

conn.commit()

print("Web Logs Table Created")
```

<h1>Create CRM Table</h1>

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS crm_conversions (
    conversion_id TEXT,
    user_id TEXT,
    lead_date TEXT,
    conversion_date TEXT,
    lead_stage TEXT,
    customer_journey TEXT,
    product TEXT,
    deal_value REAL,
    country TEXT,
    city TEXT,
    conversion_status TEXT
)
""")

conn.commit()

print("CRM Table Created")
```

<h1>Verify Tables</h1>

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table'
"""

pd.read_sql(query, conn)
```

<h1>Load CSV Data</h1>

```python
ad_spend = pd.read_csv(
    "dataset/ad_spend_clean.csv"
)

web_logs = pd.read_csv(
    "dataset/web_logs_clean.csv"
)

crm = pd.read_csv(
    "dataset/crm_clean.csv"
)
```

<h1>Check Shapes</h1>

```python
print(ad_spend.shape)
print(web_logs.shape)
print(crm.shape)
```

<h1>Load Data into SQLite</h1>

```python
ad_spend.to_sql(
    "ad_spend",
    conn,
    if_exists="replace",
    index=False
)

web_logs.to_sql(
    "web_logs",
    conn,
    if_exists="replace",
    index=False
)

crm.to_sql(
    "crm_conversions",
    conn,
    if_exists="replace",
    index=False
)

print("Data Loaded Successfully")
```
