```python
query = """
SELECT
    user_id,
    traffic_source,
    session_date,city,
    ROW_NUMBER() OVER (
        PARTITION BY user_id
        ORDER BY session_date
    ) AS step
FROM web_logs
"""
journey = pd.read_sql(query, conn)
journey.head(20)
```

```python
first_touch_query = """
SELECT
    user_id,
    traffic_source AS first_touch_channel
FROM (
    SELECT
        user_id,
        traffic_source,
        ROW_NUMBER() OVER (
            PARTITION BY user_id
            ORDER BY session_date) AS step
    FROM web_logs)
WHERE step = 1
"""
first_touch = pd.read_sql(first_touch_query, conn)
first_touch.head(20)
```

```python
first_touch_conv = first_touch.merge(
    crm,
    on="user_id",
    how="inner")
first_touch_conv.head()
```
